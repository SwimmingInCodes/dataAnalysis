```python
# https://medium.com/@kvnamipara/a-better-visualisation-of-pie-charts-by-matplotlib-935b7667d77f

import matplotlib.pyplot as plt
import seaborn as sns

# sns.get_dataset_names()

df = sns.load_dataset("titanic")
df.head()
```

```python
# counts the number of passengers for each class
a = df[df.pclass == 1].count()
print(a)
```

```python
a = df[df.pclass == 1]["pclass"].count()
print(a)
```

```python
a = df[df.pclass == 1]["pclass"].count()
b = df[df.pclass == 2]["pclass"].count()
c = df[df.pclass == 3]["pclass"].count()

# data to plot
labels = 'First', 'Second', 'Third'
sizes = [a, b, c]

explode=(0, 0, 0.1)

fig, ax=plt.subplots()

ax.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode, shadow=True, startangle=90)
ax.set_title('Titanic passengers by class', fontsize=20)

# patches, texts = plt.pie(sizes)
# print(patches)
# print(texts)

plt.legend(labels, loc="lower right", frameon=False)

# produces a perfectly circular chart
ax.axis('equal')

plt.tight_layout()
plt.show()
```

## Improvement of Pie Chart 
1. Always use good colors (preferably light shades)
2. Change labels, percent label color
3. Change positions of labels and percent labels
4. Change your pie chart to donut chart(optional)

### Change of Color
1. Hex value (#_ _ _ _ _ _)
2. rgb value ( r , g , b ) or ( r , g , b , a ) where r, g, b, a are in range 0–1
3. grey shades (0–1 range)
4. as in active color cycle ( “c0”, “c1”)[capital C followed by digit]

```python
labels = 'First', 'Second', 'Third'
sizes = [a, b, c]

explode=(0, 0, 0.1)

colors= ['#66CCFF', '#00CCFF', '#0066FF']

fig, ax=plt.subplots(dpi=100)

ax.pie(sizes, labels=labels, autopct='%1.1f%%', 
       explode=explode, colors=colors,shadow=True, startangle=90)
ax.set_title('Titanic passengers by class', fontsize=20)


plt.legend(labels, loc="lower right", frameon=False)

ax.axis('equal')

plt.tight_layout()
plt.show()
```

## Changing the color of labels on the chart

```python
labels = 'First', 'Second', 'Third'
sizes = [a, b, c]

explode=(0, 0, 0.1)

colors= ['#66CCFF', '#00CCFF', '#0066FF']

fig, ax=plt.subplots(dpi=100)

patches, texts, autotexts =ax.pie(sizes, labels=labels, autopct='%1.1f%%', 
       explode=explode, colors=colors, shadow=True, startangle=90)

ax.set_title('Titanic passengers by class', fontsize=20)

for text in texts:
    text.set_color('grey')
    text.set_size('15')
    
for autotext in autotexts:
    autotext.set_color('#FFFFFF')
    text.set_size('15') 


plt.legend(labels, loc="lower right", frameon=False)

ax.axis('equal')

plt.tight_layout()
plt.show()
```

## Change the Pie Chart to Donut Chart

```python
labels = 'First', 'Second', 'Third'
sizes = [a, b, c]

# explode=(0, 0, 0.1)

colors= ['#66CCFF', '#00CCFF', '#0066FF']

fig, ax=plt.subplots(dpi=100)

patches, texts, autotexts =ax.pie(sizes, labels=labels, autopct='%1.1f%%', 
        colors=colors, shadow=True, startangle=90)

ax.set_title('Titanic passengers by class', fontsize=20)

for text in texts:
    text.set_color('grey')
    text.set_size('15')
    
for autotext in autotexts:
    autotext.set_color('#FFFFFF')
    text.set_size('15') 

center_circle = plt.Circle((0,0), 0.60, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

plt.legend(labels, loc="lower right", frameon=False)

ax.axis('equal')

plt.tight_layout()
plt.show()
```

## Changing Positions of Labels

```python
labels = 'First', 'Second', 'Third'
sizes = [a, b, c]

explode=(0.05, 0.05, 0.05)

colors= ['#66CCFF', '#00CCFF', '#0066FF']

fig, ax=plt.subplots(dpi=100)

patches, texts, autotexts =ax.pie(sizes, labels=labels, autopct='%1.1f%%', 
        colors=colors, explode=explode, pctdistance=0.8, shadow=True, startangle=90)

ax.set_title('Titanic passengers by class', fontsize=20)

for text in texts:
    text.set_color('grey')
    text.set_size('15')
    
for autotext in autotexts:
    autotext.set_color('#FFFFFF')
    text.set_size('15') 

center_circle = plt.Circle((0,0), 0.60, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

plt.legend(labels, loc="lower right", frameon=False)

ax.axis('equal')

plt.tight_layout()
plt.show()
```

## Better Visualization of Multiple Pie Charts

```python
import pandas as pd

df=pd.DataFrame({'Language':['Python', 'C++', 'Ruby', 'Java', 'Total'],
             'Male':[315, 125, 270, 190, 900],
             'Female':[189, 212, 145, 90, 636]})
df
```

```python
df['Total']=df['Male']+df['Female']
```

```python
df.columns=df.columns.str.upper()
df
```

```python
fig, ax=plt.subplots(1, 2, dpi=120)
fig.subplots_adjust(wspace=0.4)
sizes=df['TOTAL'].loc[range(4)]
labels=df['LANGUAGE'].loc[range(4)]

colors= ['#66CCFF', '#00CCFF', '#0066FF', '#00FFFF']

ax[0].pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

sizes1=df.loc[4, 'MALE':'FEMALE']
labels1=df.columns[1:3]

ax[1].pie(sizes1, labels=labels1, colors=colors[:3], autopct='%1.1f%%', startangle=90)
```
