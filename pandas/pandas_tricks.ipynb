### Convert strings to number
- `astype()` method
- `to_numeric()` method

```python
df = pd.DataFrame({ 'product': ['A','B','C','D'], 
                   'price': ['10','20','30','40'],
                   'sales': ['20','-','60','-']
                  })
df
df.dtypes
```

```python
df['price']=df['price'].astype('int')
df
df.dtypes
```

```python
df=df.astype({'price':'int'})
```

```python
df['sales']=pd.to_numeric(df['sales'], errors='coerce')
df
```

### Detect and handle missing values

```python
titanic.info()
```

```python
titanic.isnull().sum()
titanic.isnull().sum().sum()
```

```python
titanic.isnull().mean()
```

**Dropping missing values**

```python
# To drop rows if any NaN values are present
titanic_dropped_row=titanic.dropna(axis=0)
titanic_dropped_row.isnull().sum()
```

```python
# To drop columns if any NaN values are present
titanic_dropped_column=titanic.dropna(axis=1)
titanic_dropped_column.isnull().sum()
```

```python
# To drop columns in which more than 10% of values are missing

titanic_dropped_column_th90=titanic.dropna(thresh=len(df)*0.9, axis=1)
titanic_dropped_column_th90
print(len(titanic.columns))
print(len(titanic_dropped_column_th90.columns))
```

**Replacing missing values**

```python
titanic.fillna(value=10)
titanic.fillna(axis=0, method='ffill') # 0 is rowwise filling
titanic.fillna(axis=1, method='ffill') # 1 is columnwise filling
titanic.fillna(axis=0, method='bfill')
titanic.fillna(axis=1, method='bfill')

```

```python
titanic['Age'].isnull().sum()

titanic['Age']=titanic['Age'].fillna(titanic['Age'].mean())

titanic['Age'].isnull().sum()
```

### Convert a continuous numerical feature into a categorical feature

```python
import sys


titanic['ageGroup']=pd.cut(titanic['Age'],
                          bins=[0, 13, 19, 61, sys.maxsize],
                          labels=['<12', 'Teens', 'Adult', 'Older'])

titanic.head()
```
