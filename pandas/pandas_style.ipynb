```python
# %load libs.py
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime, date, time

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity='all'

%config InlineBackend.figure_format='svg'
# plt.rcParams['figure.dpi']=120

# pd.options.display.float_format='{:,.2f}'.format
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_excel('2018_Sales_Total.xlsx')
df.head()
df.info()
```

```python
df.groupby(['name'])['ext price'].agg(['mean', 'sum'])
```

```python
(df.groupby('name')['ext price']
 .agg(['mean', 'sum'])
 .style.format('${:,.2f}'))
```

```python
(df.groupby('name')['ext price']
 .agg(['mean', 'sum'])
 .style.format('${:,.0f}'))
```

```python
# If we want to look at total sales by each month, we can use the grouper to summarize by month and 
# also calculate how much each month is as a percentage of the total annual sales.

monthly_sales = df.groupby(pd.Grouper(key='date', freq='M'))['ext price'].agg(['sum'])
monthly_sales
```

```python
monthly_sales['pct_of_total'] = monthly_sales['sum'] / df['ext price'].sum()
monthly_sales
```

```python
# Reset index: index is conveted to numrice foramt starting with 0

# Create a new dataframe, monthly_sales.
monthly_sales = df.groupby(pd.Grouper(key='date', freq='M'))['ext price'].agg(['sum']).reset_index()
monthly_sales
```

```python
monthly_sales['pct_of_total'] = monthly_sales['sum'] / df['ext price'].sum()
monthly_sales
```

```python
# Use a dictionary to define a unique formatting string for each column

format_dict = {'sum':'${:,.0f}', 'date': '{:%Y-%m-%d}', 'pct_of_total': '{:.2%}'}
monthly_sales.style.format(format_dict).hide(axis='index')
```

```python
# Highlight th highest number in green and the lowest number in color Trinidad (#cd4f39)

(monthly_sales
 .style
 .format(format_dict)
 .hide(axis='index')
 .highlight_max(color='lightgreen')
 .highlight_min(color='#cd4f39'))
```

```python
# Highlight th highest number in green and the lowest number in color Trinidad (#cd4f39)
monthly_sales_1=monthly_sales.set_index('date')
monthly_sales_2=(monthly_sales_1
 .style
 .format(format_dict)
 .highlight_max(color='lightgreen')
 .highlight_min(color='#cd4f39'))

monthly_sales_2
```

```python
# Highlight th highest number in green and the lowest number in color Trinidad (#cd4f39)

monthly_sales_3=(monthly_sales[['sum', 'pct_of_total']]
 .style
 .format(format_dict)
 .hide(axis='index')
 .highlight_max(color='lightgreen')
 .highlight_min(color='#cd4f39'))

monthly_sales_3
```

```python
(monthly_sales.style
 .format(format_dict)
 .background_gradient(subset=['sum'], cmap='BuGn'))
```
