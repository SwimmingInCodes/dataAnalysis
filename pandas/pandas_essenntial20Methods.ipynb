## Data Clenaing

**dropna()**

* dropna() method removes all the rows within the dataset.
* inplace argument gives us the option of whether we permanently drop the null values or not.
* Note: Without inplace argument, we can also assign a variable to create new data frame for example df1 = df.dropna()

````python
df.dropna(inplace=True) 
````

**fillna()**

````python
mean = df.Income.mean() # average of Income

df['Income'].fillna(mean,inplace=True)
# or df['Income'] = df['Income'].fillna(mean)
````

**drop_duplicates()**

````python
df.drop_duplicates(inplace=True) # inplace = True means we removed permanently.
````

**to_datetime()**

to_datetime() function converts an object to datetime format.

````python
df['Dt_Customer'] =  pd.to_datetime(df['Dt_Customer'])

df['year_Customer'] =  df['Dt_Customer'].dt.year
df['month_Customer'] = df['Dt_Customer'].dt.month ## month of the customers
df['month_name_Customer'] = df['Dt_Customer'].dt.month_name()
df['day_Customer'] = df['Dt_Customer'].dt.dayofweek ## dayofweek as a number 0 = Monday  6 = Sunday
# df['Dt_Customer'].dt.strftime('%A') # This shows full weekday name
````

````python
df['day_name_Customer'] = df['Dt_Customer'].dt.day_name()
````

```python

```

**groupby()**

groupby() method aggregates values and return a single scalar value according to the splitted groups.

````python
df.groupby('Education')['Income'].mean()
````

````python
df.groupby('Education').\
agg({'Income' : 'mean','Response' : 'sum','ID' : 'count'})
````

````python
df.groupby('Education').\
agg({'Income' : 'mean','Response' : 'sum','ID' : 'count'}).\
set_axis(['Average Income','# of Yes Response','# of Customers'], axis=1)
````

**pivot_table()**

````python
pd.pivot_table(data = df, values= 'Income', index = 'Education', columns = 'Marital_Status')
````

````python
df.pivot_table(values= 'Income', index = 'Education', columns = 'Marital_Status')
````

**apply()**
