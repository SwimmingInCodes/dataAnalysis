## Create a function that can calculate PPV and NPV

````python
import tkinter as tk

# Create a base
window=tk.Tk()

# Add title to window
window.title('Calculation App for Covid Diagnosis Kit Prediction Rate')

# Change the window size
window.geometry('600x600+100+200')

#############################################################
######################### Functions #########################
#############################################################
def make_lab03():
    lab03.configure(text=ent01.get()) # input text into lab03
    
def make_lab13():
    lab13.configure(text=ent11.get()) # input text into lab13
    
def make_lab23():
    lab23.configure(text=ent21.get()) # input text into lab23
    
def calculate_PPV_NPV():
    sensitivity=float(lab03.cget('text'))/100
    specificity=float(lab13.cget('text'))/100
    prevalenceRate=float(lab23.cget('text'))/100
    
    PPV=(prevalenceRate * sensitivity)/(prevalenceRate * sensitivity + (1-prevalenceRate)*(1-specificity)) * 100
    NPV=((1-prevalenceRate)*specificity)/((1-prevalenceRate)*(specificity) + prevalenceRate*(1-sensitivity)) * 100
    
    PPV=round(PPV, 2)
    NPV=round(NPV, 2)
    
    lab41.configure(text=PPV) 
    lab51.configure(text=NPV) 


#############################################################
########################## 1st Row ##########################
#############################################################

######################### Create a label ####################
lab00=tk.Label(window, 
                 text='Sensitivity', 
                 bg='#4BC55A', 
                 fg='white',
                width=8, 
                height=1,
                font=('Arial', 16, 'bold'))
lab00.grid(row=0, column=0, padx=5, pady=10)
################## Create text input window ################## 
ent01=tk.Entry(bg='white', 
               width=8, 
               font=('Arial', 16, 'bold'))
ent01.grid(row = 0, column=1, padx=5, pady=10)
####################### Create a button ###################### 
but02=tk.Button(window, 
                text='\u2192', 
                font=('Arial', 11, 'bold'), 
                bg='#0066FF', 
                fg='white', 
                width=4, 
                height=1,
               command=make_lab03)
but02.grid(row=0, column=2, padx=5, pady=10)
###################### Create another label #########################
lab03=tk.Label(window, 
                 text='', 
                 bg='#4BC55A', 
                 fg='white',
                width=8, 
                height=1,
                font=('Arial', 16, 'bold'))
lab03.grid(row=0, column=3, padx=5, pady=10)


#############################################################
########################## 2nd Row ##########################
#############################################################

######################### Create a label ####################
lab10=tk.Label(window, 
                 text='Specificity', 
                 bg='#4BC55A', 
                 fg='white',
                width=8, 
                height=1,
                font=('Arial', 16, 'bold'))
lab10.grid(row=1, column=0, padx=5, pady=10)
################## Create text input window ################## 
ent11=tk.Entry(bg='white', 
               width=8, 
               font=('Arial', 16, 'bold'))
ent11.grid(row = 1, column=1, padx=5, pady=10)
####################### Create a button ###################### 
but12=tk.Button(window, 
                text='\u2192', 
                font=('Arial', 11, 'bold'), 
                bg='#0066FF', 
                fg='white', 
                width=4, 
                height=1,
               command=make_lab13)
but12.grid(row=1, column=2, padx=5, pady=10)
###################### Create another label #########################
lab13=tk.Label(window, 
                 text='', 
                 bg='#4BC55A', 
                 fg='white',
                width=8, 
                height=1,
                font=('Arial', 16, 'bold'))
lab13.grid(row=1, column=3, padx=5, pady=10)


#############################################################
########################## 3rd Row ##########################
#############################################################

######################### Create a label ####################
lab20=tk.Label(window, 
                 text='Specificity', 
                 bg='#4BC55A', 
                 fg='white',
                width=8, 
                height=1,
                font=('Arial', 16, 'bold'))
lab20.grid(row=2, column=0, padx=5, pady=10)
################## Create text input window ################## 
ent21=tk.Entry(bg='white', 
               width=8, 
               font=('Arial', 16, 'bold'))
ent21.grid(row = 2, column=1, padx=5, pady=10)
####################### Create a button ###################### 
but22=tk.Button(window, 
                text='\u2192', 
                font=('Arial', 11, 'bold'), 
                bg='#0066FF', 
                fg='white', 
                width=4, 
                height=1,
               command=make_lab23)
but22.grid(row=2, column=2, padx=5, pady=10)
###################### Create another label #########################
lab23=tk.Label(window, 
                 text='', 
                 bg='#4BC55A', 
                 fg='white',
                width=8, 
                height=1,
                font=('Arial', 16, 'bold'))
lab23.grid(row=2, column=3, padx=5, pady=10)


#############################################################
########################## 4th Row ##########################
#############################################################

but30=tk.Button(window, 
                text='Calculation (click)', 
                font=('Arial', 16, 'bold'), 
                bg='#0066FF', 
                fg='white', 
                width=16, 
                height=1,
               command=calculate_PPV_NPV)
but30.grid(row=3, column=0, padx=5, pady=10, columnspan=4, sticky='we') # create button automatically expand from west to east


#############################################################
########################## 5th Row ##########################
#############################################################

lab40=tk.Label(window, 
                 text='Predictive Positive Value', 
                 bg='#4BC55A', 
                 fg='white',
                width=8, 
                height=1,
                font=('Arial', 16, 'bold'))
lab40.grid(row=4, column=0, padx=5, pady=10, columnspan=3, sticky='we')


lab41=tk.Label(window, 
                 text='', 
                 bg='yellow', 
                 fg='black',
                width=8, 
                height=1,
                font=('Arial', 16, 'bold'))
lab41.grid(row=4, column=3, padx=5, pady=10)


#############################################################
########################## 6th Row ##########################
#############################################################

lab50=tk.Label(window, 
                 text='Predictive Negative Value', 
                 bg='#4BC55A', 
                 fg='white',
                width=8, 
                height=1,
                font=('Arial', 16, 'bold'))
lab50.grid(row=5, column=0, padx=5, pady=10, columnspan=3, sticky='we')


lab51=tk.Label(window, 
                 text='', 
                 bg='yellow', 
                 fg='black',
                width=8, 
                height=1,
                font=('Arial', 16, 'bold'))
lab51.grid(row=5, column=3, padx=5, pady=10)


window.mainloop()
````
