## Fill Between Lines

```python
%matplotlib inline
%config InlineBackend.figure_format='svg'
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('data_programmer.csv')
df.head()
```

```python
ages=df['Age']
dev_salaries=df['All_Devs']
py_salaries=df['Python']
ja_salaries=df['JavaScript']

fig, ax=plt.subplots(figsize=(10, 5))
ax.plot(ages, dev_salaries, linestyle='--', linewidth=1, label='Dev salaries')
ax.plot(ages, py_salaries, linestyle='-', linewidth=1, label='Python salaries')


ax.fill_between(ages, dev_salaries, py_salaries, where=(py_salaries>dev_salaries), 
                label='Above Average', 
                color='green', 
                alpha=0.2,
               interpolate=True)

ax.fill_between(ages, dev_salaries, py_salaries, where=(py_salaries <=dev_salaries), 
                label='Below Average', 
                color='red', 
                alpha=0.2,
               interpolate=True)

ax.set_title('Median Salary (USD) by Age', fontsize=25, y=1.02)
ax.set_xlabel('Ages', fontsize=15)
ax.set_ylabel('Median Salary (USD)', fontsize=15)

for label in ax.get_xticklabels():
    label.set_fontsize(15)

ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.yaxis.set_ticks_position('right')

leg=ax.legend(frameon=False, labelspacing=0.5, borderpad=1)
leg.set_title('Salaries', prop={'size': 12})

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('data_programmer.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig1, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)


ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
