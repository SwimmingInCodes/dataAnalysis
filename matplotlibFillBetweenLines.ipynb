## Fill Between Lines

```python
%matplotlib inline
%config InlineBackend.figure_format='svg'
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('data_programmer.csv')
df.head()
```

```python
ages=df['Age']
dev_salaries=df['All_Devs']
py_salaries=df['Python']
ja_salaries=df['JavaScript']

fig, ax=plt.subplots(figsize=(10, 5))
ax.plot(ages, dev_salaries, linestyle='--', linewidth=1, label='Dev salaries')
ax.plot(ages, py_salaries, linestyle='-', linewidth=1, label='Python salaries')

leg=ax.legend(frameon=False, labelspacing=0.5, borderpad=1)
leg.set_title('Salaries', prop={'size': 12})

ax.fill_between(ages, dev_salaries, py_salaries, where=(py_salaries>dev_salaries), color='green', alpha=0.2,
               interpolate=True)

ax.fill_between(ages, dev_salaries, py_salaries, where=(py_salaries <=dev_salaries), color='red', alpha=0.2,
               interpolate=True)

ax.set_title('Median Salary (USD) by Age', fontsize=25, y=1.02)
ax.set_xlabel('Ages', fontsize=15)
ax.set_ylabel('Median Salary (USD)', fontsize=15)

for label in ax.get_xticklabels():
    label.set_fontsize(15)

ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.yaxis.set_ticks_position('right')

plt.tight_layout()
plt.show()
```
