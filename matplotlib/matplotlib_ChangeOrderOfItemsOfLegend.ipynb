## Change order of items in the legend

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax=plt.subplots(figsize=(8,6), dpi=80)

ht=ax.add_patch(mpatches.Rectangle((1, 1), 1, 1, color='r', label= 'Top', alpha=.5))
h1=ax.bar(1, 2, label='Middle')
hb=ax.add_patch(mpatches.Rectangle((0.5, 0.0), 1, 1, color='k', label='Bottom', alpha=.5))
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax=plt.subplots(figsize=(8,6), dpi=80)

ht=ax.add_patch(mpatches.Rectangle((1, 1), 1, 1, color='r', label= 'Top', alpha=.5)) # (x_point, y_point), width, height
h1=ax.bar(1, 2, label='Middle')
hb=ax.add_patch(mpatches.Rectangle((0.5, 0.0), 1, 1, color='k', label='Bottom', alpha=.5))

handles, labels=plt.gca().get_legend_handles_labels()
print(handles, '\n')
print(labels)

order=[2, 1, 0]
plt.legend([handles[idx] for idx in order], [labels[idx] for idx in order])
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax=plt.subplots(figsize=(8,6), dpi=80)

ht=ax.add_patch(mpatches.Rectangle((1, 1), 1, 1, color='r', label= 'Top', alpha=.5)) # (x_point, y_point), width, height
h1=ax.bar(1, 2, label='Middle')
hb=ax.add_patch(mpatches.Rectangle((0.5, 0.0), 1, 1, color='k', label='Bottom', alpha=.5))

handles, labels=plt.gca().get_legend_handles_labels()
print(handles, '\n')
print(labels)

order=[2, 1, 0]

leg=plt.legend([handles[idx] for idx in order], [labels[idx] for idx in order], frameon=False, fontsize= 12, 
           labelspacing=0.5, borderpad=1, loc=(1, 0.75))
leg.set_title("Location", prop={'size':15})

plt.show()
```
