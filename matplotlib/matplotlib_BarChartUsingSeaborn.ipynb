## Create Bar Charts Using Searborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# list up all the datasetes inlcuded in seaborn
sns.get_dataset_names()
```

```python
crashes=sns.load_dataset("car_crashes")
df=crashes.loc[range(5)]
df.head()
```

## <span style="color:red"> "To avoid No handles with labels found to put in legend", </span> we need to assign each barplot to a variable
As shown below, the first barplot is assgined to "total", and the second one is to "speed"

```python
plt.rcParams["figure.dpi"]=120
fig, ax=plt.subplots()

sns.set_color_codes("pastel")
total=sns.barplot(x='total', y='abbrev', data=df, label="Total",  color='b')
sns.set_color_codes("muted")
speed=sns.barplot(x='speeding', y='abbrev', data=df, label="Speed-related",  color='b')

plt.title("Speeding-related automobile collisions", fontsize=15)
plt.legend(ncol=1, loc="lower right", )
ax.set(xlim=(0,30), ylabel="State", xlabel="Automobile collisions (per billion miles)")
plt.show()
```

## Sort the bars from largest total number to the smallest one
Apply sort_values() method to "total" column with ascending to False

```python
plt.rcParams["figure.dpi"]=120
df=crashes.loc[range(5)]

df1=df.sort_values('total', ascending=False)

sns.set_color_codes("pastel")
total=sns.barplot(x='total', y='abbrev', data=df1, label="Total",  color='b')
sns.set_color_codes("muted")
speed=sns.barplot(x='speeding', y='abbrev', data=df1, label="Speed-related",  color='b')

plt.title("Speeding-related automobile collisions", fontsize=15)
plt.legend(ncol=1, loc="lower right", )

plt.show()
```

## Add the values of the totals to the outside of the bars
### [ref](https://www.delftstack.com/howto/matplotlib/add-value-labels-on-matplotlib-bar-chart/#:~:text=To%20add%20value%20labels%20on%20a%20Matplotlib%20bar%20chart%2C%20we,any%20location%20in%20the%20graph.) 

```python
plt.rcParams["figure.dpi"]=150
df=crashes.loc[range(5)]
df1=df.sort_values('total', ascending=False)

def add_value(x_list, y_list):
    for i in range(len(y_list)):
        plt.annotate(x_list[i], (x_list[i], i)) # the first argument is "text" and the second one is xy points
        
# print(ax.get_xticklabels())
# print(ax.get_yticklabels()[1])
# print(df1.total.to_list())

x_list=df1["total"].to_list() 
print(x_list)

y_list=ax.get_yticks()
print(y_list)

sns.set_color_codes("pastel")
total=sns.barplot(x='total', y='abbrev', data=df1, label="Total",  color='b')
sns.set_color_codes("muted")
speed=sns.barplot(x='speeding', y='abbrev', data=df1, label="Speed-related",  color='b')

total.set_xlim([0, 25])
total.set_ylabel("State", fontsize=12)
total.set_xlabel("Automobile collisions (per billion miles)", fontsize=12)
add_value (x_list, y_list)       

plt.title("Speeding-related automobile collisions", fontsize=15)
plt.legend(ncol=1, loc="lower right", )
plt.show()
```

## Add the values of totals inside of the bars
