## Analysis the top 200 trending videos on YouTube using scatter plot

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('2019-05-31-data.csv')
data.head()
```

```python
%config InlineBackend.figure_formats=['svg']

view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.tight_layout()
plt.show()
```

### In the above graph, we can find one point that is located far away from most of points but that is a meaningful point so, in this case, we can use logscale to lessen the outlier to skew the plot. 

```python
%config InlineBackend.figure_formats=['svg']

view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.tight_layout()
plt.show()
```

## Analysis of population distribution of CA using scatter plot

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('california_cities.csv')
df.head()
```

```python
%matplotlib inline
%config InlineBackend.figure_format='svg'

lat, long=df['latd'], df['longd']
area, population=df['area_total_sq_mi'], df['population_total']

fig, ax=plt.subplots(figsize=(12, 6))
ax.scatter(lat, long, c=np.log10(population), s=area, cmap='viridis')

cmap=plt.get_cmap('viridis', 20)
norm=mpl.colors.Normalize(vmin=0, vmax=2)
sm=plt.cm.ScalarMappable(cmap=cmap, norm=norm)
plt.colorbar(sm, ticks=np.linspace(0, 2, 20), boundaries=np.arange(0.05, 2.1, 0.1), label='log$_{10}$(population)')

for i in [100, 300, 500]:
    ax.scatter([], [], s=i, c='k', alpha=0.2, label=str(i)+ ' km$^2$')
    
leg=ax.legend(scatterpoints=1, frameon=False, fontsize= 12, labelspacing=1, borderpad=2)
leg.set_title('City Area', prop={'size': 20})

for label in ax.get_xticklabels():
    label.set_ha('center')
    label.set_fontsize(15)
#     label.set_rotation(45)

for label in ax.get_yticklabels():
    label.set_fontsize(15)
    
ax.set_xlabel('Latitude', fontsize=20)
ax.set_ylabel('Longitude', fontsize=20)
ax.yaxis.set_label_coords(-0.1, 0.5)
    
plt.suptitle('\n'.join(['Populatoin of City', 'California', '2021 Census']), fontsize=20,x=0.45, y=1.02)
```
