**Create user own contextmanager for file I/O**

```python
class filemanager:
    def __init__(self, filename, mode):
        self.filename=filename
        self.mode=mode
        
    def __enter__(self):
        self.file=open(self.filename, self.mode)
        return self.file       
    
    def __exit__(self, exval, extype, traceback): # exit with exception related arguments
        return self.file.close()      
```

```python
with filemanager('./data/text1.txt', 'w') as f:
    f.write('This is your book')
```

**copy the image with chunks**

```python
with open('./data/Chihuahua.png', 'rb') as rf:
    with open('Chihuahua_copy2.png', 'wb') as wf:
        size_to_read=4096
        content=rf.read(size_to_read)
        while len(content)>0:
            wf.write(content)
            content=rf.read(size_to_read)      
```
